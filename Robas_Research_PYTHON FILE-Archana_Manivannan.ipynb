**Sentiment Analysis and Word Cloud Generation**

We'll analyze open-ended columns from your survey:

Q11_NeighborhoodFeedback

Q12_FutureConcerns

```python
import pandas as pd
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt')  # For TextBlob tokenization
```

```python
df.info()
```

```python
# Load your dataset
df = pd.read_csv("Synthetic_Survey.csv")

# Fill missing with empty string
df["Q11_NeighborhoodFeedback"] = df["Q11_OE_Neighborhood"].fillna("")
df["Q12_FutureConcerns"] = df["Q12_OE_FutureConcerns"].fillna("")
```

```python
# Sentiment function
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
# Apply sentiment analysis
df["Neighborhood_Sentiment"] = df["Q11_NeighborhoodFeedback"].apply(get_sentiment)
df["FutureConcern_Sentiment"] = df["Q12_FutureConcerns"].apply(get_sentiment)
```

```python
# Summary statistics
print("\nNeighborhood Sentiment Summary:\n", df["Neighborhood_Sentiment"].describe())
print("\nFuture Concern Sentiment Summary:\n", df["FutureConcern_Sentiment"].describe())
```

```python
# Plot sentiment histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df["Neighborhood_Sentiment"], bins=20, color="skyblue", edgecolor="black")
plt.title("Sentiment of Neighborhood Feedback")
plt.xlabel("Polarity Score")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(df["FutureConcern_Sentiment"], bins=20, color="salmon", edgecolor="black")
plt.title("Sentiment of Future Concerns")
plt.xlabel("Polarity Score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Generate word clouds
text1 = " ".join(df["Q11_NeighborhoodFeedback"])
text2 = " ".join(df["Q12_FutureConcerns"])

wordcloud1 = WordCloud(width=800, height=400, background_color='white').generate(text1)
wordcloud2 = WordCloud(width=800, height=400, background_color='white').generate(text2)
```

```python
# Display word clouds
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud: Neighborhood Feedback")

plt.subplot(1, 2, 2)
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud: Future Concerns")

plt.tight_layout()
plt.show()
```

**Export cleaned file for Power BI**

```python
import pandas as pd
from textblob import TextBlob

# Load data
df = pd.read_csv("Synthetic_Survey.csv")

# Fix typos
df['Q3_Education'] = df['Q3_Education'].replace({
    "Primry": "Primary", "Secndary": "Secondary", "Tertary": "Tertiary"
})
df['Q8_MaritalStatus'] = df['Q8_MaritalStatus'].replace({
    "Widowd": "Widowed", "Divorcd": "Divorced", "Singel": "Single", "Marrid": "Married"
})

# Fill missing open-ended responses
df["Q11_OE_Neighborhood"] = df["Q11_OE_Neighborhood"].fillna("")
df["Q12_OE_FutureConcerns"] = df["Q12_OE_FutureConcerns"].fillna("")

# Sentiment analysis
df["Neighborhood_Sentiment"] = df["Q11_OE_Neighborhood"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["FutureConcern_Sentiment"] = df["Q12_OE_FutureConcerns"].apply(lambda x: TextBlob(x).sentiment.polarity)

# Export CSV
df.to_csv("Survey_Insights_for_PowerBI.csv", index=False)
```
